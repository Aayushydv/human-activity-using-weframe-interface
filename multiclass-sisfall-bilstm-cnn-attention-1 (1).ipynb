```python
cnn_model = tf.keras.models.Sequential()

#First CNN layer  with 32 filters, conv window 3, relu activation and same padding
cnn_model.add(Conv1D(filters=32, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001), input_shape = (9,1,)))

#Second CNN layer  with 64 filters, conv window 3, relu activation and same padding
cnn_model.add(Conv1D(filters=64, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

#Third CNN layer with 128 filters, conv window 3, relu activation and same padding
cnn_model.add(Conv1D(filters=128, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

#Fourth CNN layer with Max pooling
cnn_model.add(MaxPool1D(pool_size=(3,), strides=2, padding='same'))
cnn_model.add(Dropout(0.5))

#Flatten the output
cnn_model.add(Flatten())

#Add a dense layer with 256 neurons
#cnn_model.add(Dense(units = 256, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

#Add a dense layer with 512 neurons
#cnn_model.add(Dense(units = 512, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

#Softmax as last layer with two outputs
cnn_model.add(Dense(units = 8, activation='softmax'))
cnn_model.summary()
```

```python
# Compile the model
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.models import load_model, Sequential

# Assuming you have an existing model named cnn_model
cnn_model = load_model('cnn_model.h5')

# Create a new model with the same architecture
cnn_attn_model = Sequential()
for layer in cnn_model.layers:
    cnn_attn_model.add(layer)

# Compile the new model with the same configurations
cnn_attn_model.compile(optimizer=cnn_model.optimizer,
                       loss=cnn_model.loss,
                       metrics=cnn_model.metrics)

# Optionally, you can save the duplicate model to a new file
cnn_attn_model.save('cnn_attn_model.h5')
```

```python
from keras.layers import Layer, Dense
import tensorflow as tf

# Custom layer for channel attention
class ChannelAttention(Layer):
    def __init__(self, **kwargs):
        super(ChannelAttention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.channels = input_shape[-1]
        self.dense_1 = Dense(self.channels//2, activation='relu')
        self.dense_2 = Dense(self.channels, activation='sigmoid')
        super(ChannelAttention, self).build(input_shape)

    def call(self, inputs):
        avg_pool = tf.reduce_mean(inputs, axis=1, keepdims=True)
        avg_pool = self.dense_1(avg_pool)
        max_pool = tf.reduce_max(inputs, axis=1, keepdims=True)
        max_pool = self.dense_1(max_pool)
        channel_attention = tf.nn.sigmoid(self.dense_2(tf.concat([avg_pool, max_pool], axis=-1)))
        return inputs * channel_attention

# ...


# Add channel attention to the model using the custom layer
cnn_attn_model.add(ChannelAttention())

# Compile the model
cnn_attn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with channel attention
history_attn = cnn_attn_model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_val, y_val))

# Use the model for evaluation
test_loss_attn, test_accuracy_attn = cnn_attn_model.evaluate(x_test, y_test, verbose=2)
print(f'Test Loss with Channel Attention: {test_loss_attn:.4f}')
print(f'Test Accuracy with Channel Attention: {test_accuracy_attn:.4f}')

# Save the models
cnn_attn_model.save("cnn_attn_model.h5")
cnn_model.save("cnn_model.h5")
```

```python
class TemporalAttention(Layer):
    def __init__(self, **kwargs):
        super(TemporalAttention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.time_steps = input_shape[1]
        self.dense_1 = Dense(self.time_steps//2, activation='relu')
        self.dense_2 = Dense(self.time_steps, activation='sigmoid')
        super(TemporalAttention, self).build(input_shape)

    def call(self, inputs):
        avg_time = tf.reduce_mean(inputs, axis=-1, keepdims=True)
        avg_time = self.dense_1(avg_time)
        max_time = tf.reduce_max(inputs, axis=-1, keepdims=True)
        max_time = self.dense_1(max_time)
        temporal_attention = tf.nn.sigmoid(self.dense_2(tf.concat([avg_time, max_time], axis=1)))
        return inputs * temporal_attention

# ... (previous code)

# Create a new model for the temporal attention
cnn_temporal_model = tf.keras.models.Sequential()

# First CNN layer with 32 filters, conv window 3, relu activation and same padding
cnn_temporal_model.add(Conv1D(filters=32, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001), input_shape=(9, 1,)))

# Second CNN layer with 64 filters, conv window 3, relu activation and same padding
cnn_temporal_model.add(Conv1D(filters=64, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

# Third CNN layer with 128 filters, conv window 3, relu activation and same padding
cnn_temporal_model.add(Conv1D(filters=128, kernel_size=(3,), padding='same', activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

# Fourth CNN layer with Max pooling
cnn_temporal_model.add(MaxPool1D(pool_size=(3,), strides=2, padding='same'))
cnn_temporal_model.add(Dropout(0.5))

# Flatten the output
cnn_temporal_model.add(Flatten())

# Add a dense layer with 256 neurons
cnn_temporal_model.add(Dense(units=256, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

# Add a dense layer with 512 neurons
cnn_temporal_model.add(Dense(units=512, activation=tf.keras.layers.LeakyReLU(alpha=0.001)))

# Softmax as the last layer with eight outputs
cnn_temporal_model.add(Dense(units=8, activation='softmax'))
```

```python
# Compile the model
cnn_temporal_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with temporal attention
history_temporal = cnn_temporal_model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_val, y_val))

# Use the model for evaluation
test_loss_temporal, test_accuracy_temporal = cnn_temporal_model.evaluate(x_test, y_test, verbose=2)
print(f'Test Loss with Temporal Attention: {test_loss_temporal:.4f}')
print(f'Test Accuracy with Temporal Attention: {test_accuracy_temporal:.4f}')

# Save the models
cnn_temporal_model.save("cnn_temporal_model.h5")
```
